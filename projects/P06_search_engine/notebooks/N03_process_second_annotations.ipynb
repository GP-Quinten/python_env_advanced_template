```python
import json
import pandas as pd
from p06_search_engine import config
```

# constants

```python
PATH_DATA = "/home/jperrio/registry_data_catalog_experiments/datasets/007_search_engine_datasets/2nd_annotations"
PATH_DATA_QUERIES = f"{PATH_DATA}/queries_GH20250725.xlsx"
PATH_DATA_ANNOTATIONS = f"{PATH_DATA}/annotations_RR20250804.xlsx" #FIXME
```

# Processing

## queries

```python
# Load queries
df_queries = (
    pd.read_excel(PATH_DATA_QUERIES, header=None)

    .rename(columns={0: "query_text"})
    .assign(query_text=lambda df: df["query_text"].str.strip())
    
    .sort_values("query_text")

    .assign(query_id=lambda df: df["query_text"].rank().astype("int").apply(lambda rank: f"2nd-{rank:03d}"))

    [["query_id", "query_text"]]
)
display(df_queries)
```

```python
# Build dataset of queries
queries = [
    row_query.to_dict()
    for _, row_query in df_queries.iterrows()
]
```

```python
# Save dataset of queries
with open(config.PATH_DATA_2ND_QUERIES, "w") as file:
    json.dump(queries, file, indent=4)
```

## annotations

```python
# Load annotations

# iterate over sheets
dfs_annotations = []
for sheet_name, df_annotations in pd.read_excel(PATH_DATA_ANNOTATIONS, sheet_name=None).items():
    if sheet_name == "Data":
        continue
    dfs_annotations.append(df_annotations.loc[lambda df: df["query_text"].notna()])

# concatenate all dataframes
df_annotations = pd.concat(dfs_annotations, ignore_index=True)

df_annotations
```

```python
fit_to_request_columns = ["fit to request_Ghinwa", "Soso fit for request", "Sonia fit for request", "fit to request_Ryane", "fit to request", "fit to request.1"]

(
    df_annotations

    .merge(df_queries, on="query_text", how="left")

    .assign(annotation_id=lambda df: df["query_text"].expanding().count())
    .assign(annotation_id=lambda df: df["annotation_id"].apply(lambda annotation_id: f"2nd-{int(annotation_id):03d}"))

    .loc[lambda df: df[fit_to_request_columns].notna().any(axis=1)]
    .assign(**{col: lambda df, col=col: df[col].str.strip() for col in fit_to_request_columns})
    .assign(**{col: lambda df, col=col: df[col].mask(df[col].eq("Maybe")) for col in fit_to_request_columns})
    .loc[lambda df: df[fit_to_request_columns].notna().any(axis=1)]
    .assign(fit_to_request=lambda df: df[fit_to_request_columns].agg(lambda s: s.dropna().iloc[0], axis=1))
    .drop(fit_to_request_columns, axis=1)

    .loc[lambda df: df["fit_to_request"].notna()]
    .assign(annotation_label=lambda df: df["fit_to_request"].replace({"Yes": "YES", "No": "NO"}))
    .loc[lambda df: df["annotation_label"].notna()]

    [["query_id", "registry_id", "annotation_id", "annotation_label"]]
)
```

```python
fit_to_request_columns = ["fit to request_Ghinwa", "Soso fit for request", "Sonia fit for request", "fit to request_Ryane", "fit to request", "fit to request.1"]

df_annotations = (
    df_annotations

    .merge(df_queries, on="query_text", how="left")

    .assign(annotation_id=lambda df: df["query_text"].expanding().count())
    .assign(annotation_id=lambda df: df["annotation_id"].apply(lambda annotation_id: f"2nd-{int(annotation_id):03d}"))

    .loc[lambda df: df[fit_to_request_columns].notna().any(axis=1)]
    .assign(**{col: lambda df, col=col: df[col].str.strip() for col in fit_to_request_columns})
    .assign(**{col: lambda df, col=col: df[col].mask(df[col].eq("Maybe")) for col in fit_to_request_columns})
    .loc[lambda df: df[fit_to_request_columns].notna().any(axis=1)]
    .assign(fit_to_request=lambda df: df[fit_to_request_columns].agg(lambda s: s.dropna().iloc[0], axis=1))
    .drop(fit_to_request_columns, axis=1)

    .loc[lambda df: df["fit_to_request"].notna()]
    .assign(annotation_label=lambda df: df["fit_to_request"].replace({"Yes": "YES", "No": "NO"}))
    .loc[lambda df: df["annotation_label"].notna()]

    [["query_id", "registry_id", "annotation_id", "annotation_label"]]
)
```

```python
display(df_annotations)

annotations = df_annotations.to_dict(orient="records")

# Save dataset of annotations
with open(config.PATH_DATA_2ND_ANNOTATIONS, "w") as file:
    json.dump(annotations, file, indent=4)
```
